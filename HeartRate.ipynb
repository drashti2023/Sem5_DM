 # 22. Age vs Max Heart Rate Scatter Plot

```python
colors = df['target'].map({0: 'blue', 1: 'red'}) 
plt.figure(figsize=(8,5))
plt.scatter(df['age'], df['thalach'], c=colors, alpha=0.6)
plt.title("Age vs Max Heart Rate (Red = Heart Disease)")
plt.xlabel("Age")
plt.ylabel("Max Heart Rate (thalach)")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
