```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
```

# Step 1: Load Data

```python
df = pd.read_csv("OnlineRetail.csv", encoding='ISO-8859-1')
df
```

# Step 2: Drop Missing Invoices or Descriptions

```python
df1 = df.dropna(subset=['InvoiceNo','Description'])
df1
```

# Step 3: Remove Canceled Invoices <BR> Start with letter 'c'

```python
df1 = df1[~df1['InvoiceNo'].astype(str).str.startswith('C')]
df1
```

# Step 4: Keep Only Positive Quantities

```python
df1 = df1[df1['Quantity'] > 0]
df1
```

```python
print((df1['Quantity'] <= 0).sum())
```

# Step 5: Standardize Item Descriptions | Lower case and remove space

```python
df1['Description'] = df1['Description'].str.lower().str.strip()
df1
```

# Step 6: Filter for a Single Country (e.g., United Kingdom)

```python
df1 = df1[df1['Country'] == 'United Kingdom']
df1
```

# Step 7: Remove Duplicates

```python
df1.drop_duplicates()
```

# Step 8: Add TotalPrice column , TotalPrice = Quantity *  UnitPrice

```python
df1['TotalPrice'] = df1['Quantity'] * df1['UnitPrice']
df1
```

# Step 9 : Check for Top 10 Selling Products

```python
top = df1.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
print(top)
```

# Step 10 : Identify Most Active Customers

```python
act = (df1.groupby('CustomerID')['InvoiceNo'].count().sort_values(ascending=False).head(10))
print(act)
```

# Step 11: Time-Based Analysis | Convert InvoiceDate to datetime and check popular months/hours.

```python
df1['InvoiceDate'] = pd.to_datetime(df1['InvoiceDate'])

df1['Month'] = df1['InvoiceDate'].dt.month

popular_months = df1['Month'].value_counts().sort_index()
print(popular_months)

df1['Hour'] = df1['InvoiceDate'].dt.hour

popular_hours = df1['Hour'].value_counts().sort_index()
print(popular_hours)
```

```python
import matplotlib.pyplot as plt

popular_months.plot(kind='bar', color='teal', figsize=(10,6), title='Transactions per Month')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

popular_hours.plot(kind='bar', color='orange', figsize=(10,6), title='Transactions by Hour of Day')
plt.ylabel('Number of Transactions')
plt.xlabel('Hour (24h)')
plt.tight_layout()
plt.show()
```
