### 20. Display top 10 products sold (by Quantity)

```python
top_10 = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
top_10.plot(kind='bar', figsize=(12, 6), color='teal', edgecolor='black')
plt.title('Top 10 Products Sold by Quantity')
plt.xlabel('Product Description')
plt.ylabel('Total Quantity Sold')
plt.show()
```

### 21. Display top 10 revenue-generating products

```python
top_10_revenue = df.groupby('Description')['TotalAmount'].sum().sort_values(ascending=False).head(10)
top_10_revenue.plot(kind='bar', figsize=(12, 6), color='limegreen', edgecolor='black')
plt.title('Top 10 Revenue-Generating Products')
plt.xlabel('Product Description')
plt.ylabel('Total Revenue')
plt.show()
```

### 22. Display number of orders per country

```python
orders_per_country = df.groupby('Country')['InvoiceNo'].nunique().sort_values(ascending=False)
orders_per_country.plot(kind='bar', figsize=(12, 6), color='purple', edgecolor='black')
plt.title('Number of Orders per Country')
plt.xlabel('Country')
plt.ylabel('Number of Unique Orders')
plt.show()
```

### 23. Display orders per hour of the day

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Hour'] = df['InvoiceDate'].dt.hour

orders_per_hour = df.groupby('Hour')['InvoiceNo'].nunique()

orders_per_hour.plot(kind='bar', figsize=(10, 5), color='tomato', edgecolor='black')
plt.title('Number of Orders per Hour of the Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Unique Orders')
plt.show()
```

### 24. Display daily sales trend (line chart)

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
daily_sales = df.groupby(df['InvoiceDate'].dt.date)['TotalAmount'].sum()

daily_sales.plot(kind='line', figsize=(12, 6), color='green')
plt.title('Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()
```

### 25. Display monthly sales trend 

```python
df['Month'] = df['InvoiceDate'].dt.to_period('M')
monthly_sales = df.groupby('Month')['TotalAmount'].sum()

monthly_sales.plot(kind='line', figsize=(12, 6), marker='o', color='blue')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```
