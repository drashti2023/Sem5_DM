### 9) Display the histogram for Age attribute by discretizing it into 8 separate bins and counting the frequency for each bin.

```python
import matplotlib.pyplot as plt
```

```python
data['Age'].hist(bins=8)
plt.xlabel("Age")
plt.ylabel("Frequency")
```

```python
plt.hist(data['Age'], bins=8, edgecolor='black')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram of Age (8 Bins)")
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()
```

### 11)  Display scatter plot for any 5 pair of attributes , we can use a scatter plot to visualize their joint distribution.

```python
plt.scatter(data['Age'],data['Fare'],alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')
plt.show()
```

```python
pairs = [('Age', 'Fare'), ('Pclass', 'Survived'), ('SibSp', 'Parch'), ('Age', 'Pclass'), ('Fare', 'Survived')]

plt.figure(figsize=(15, 10))
for idx, (x, y) in enumerate(pairs):
    plt.subplot(2, 3, idx + 1)
    plt.scatter(data[x], data[y], alpha=0.5)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title(f"{x} vs {y}")
    plt.tight_layout()
    plt.show()

```
